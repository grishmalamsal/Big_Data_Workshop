factorial of a number (user input)

```python
num= int (input("enter a number :"))
fact =1
for i in range (1, num+1):
  fact*=i

  print ("fatcorial:", fact)
```

calculator using function

```python
def add (a,b):
  return a+b
def subtract(a,b):
    return a-b
def multiply (a,b):
      return a*b
def divide (a,b):
        if b==0:
           return "cannot devide by zero"
        return a/b
 #user input
x= float(input("enter first number:"))
y= float(input("enter second number:"))
op= input("enter operation(+,-,*,/):")

if op =="+":
            print("result:", add(x,y))
elif op =="-":
            print("result:", subtract(x,y))
elif op =="*":
            print("result:", multiply(x,y))
elif op =="/":
            print("result:", divide(x,y))
else:
            print("invalid operation")

```

remove first duplicate element from list

```python
lst =[1,2,3,2,4,5]
seen =set()

for i in range (len (lst)):
  if lst [i]in seen:
      lst.pop(i)
      break
seen.add (lst[i])
print (lst)
```

extra specific keys from dictonary

```python
sample_dict={
    "name":"grishma",
    "age":22,
    "salary":10000,
    "city":"kathmandu"
}
keys=["name","salary"]
new_dict={k:sample_dict[k] for k in keys}
print(new_dict)
```
